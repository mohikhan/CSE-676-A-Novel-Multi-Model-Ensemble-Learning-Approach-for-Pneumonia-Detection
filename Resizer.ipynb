```python
import PIL
import os
import os.path
from PIL import Image
os.getcwd( )
```

```python
#resizing
# Attention TA: Please change the directory path according to your convenience wherever you will store the dataset.
# The directory path here is the path where we have store the dataset in our local machine.
img_dir = r'C:\Users\mdavo\Desktop\archive\chest_xray\train\PNEUMONIA'
print('Bulk images resizing started...')
for img in os.listdir(img_dir):
	f_img = img_dir + '\\' + img
	f, e = os.path.splitext(img_dir + img)
	img = Image.open(f_img)
	img = img.resize((100, 100))
	img.save(f + '_resized.jpg')
	print(img_dir)

print('Bulk images resizing finished...')
```
